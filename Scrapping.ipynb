```python
import time
import pandas as pd
import numpy as np
from tqdm import tqdm
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
service_obj = Service("C:\\Users\\rashp\\anaconda3\\lib\\site-packages\\chromedriver_binary\\chromedriver.exe")


browser = webdriver.Chrome(service=service_obj) 
```

```python
data = []
for page in tqdm(range(1,1001)):
    link = 'https://www.naukri.com/indian-portal-jobs-' + str(page)
    browser.get(link)
    time.sleep(2)
    soup = BeautifulSoup(browser.page_source , 'html.parser')
    for job in soup.find_all('article', class_ = 'jobTuple'):
        
        try:    
            job_id           = job.get('data-job-id')   
        except:
            job_id           = np.nan    

        try: 
            job_link         = job.find('a' , class_ = 'title ellipsis').get('href')
        except:
            job_link         = np.nan

        try:
            job_role         = job.find('a' , class_ = 'title ellipsis').text.strip()
        except:
            job_role         = np.nan

        try:
            company         = job.find('div' , class_ = 'companyInfo subheading').find('a' , class_ = 'subTitle ellipsis fleft').text.strip()
        except:
            company          = np.nan

        try:
            company_link     = job.find('div' , class_ = 'companyInfo subheading').find('a').get('href')
        except:
            company_link     = np.nan        

        try:
            rating          = job.find('div' , class_ = 'companyInfo subheading').find('span' , class_ = 'starRating fleft').text.strip()
        except:
            rating           = np.nan

        try:
            reviews          = job.find('div' , class_ = 'companyInfo subheading').find('a' , class_ = 'reviewsCount fleft').text.strip()
        except:
            reviews          = np.nan

        try:
            experience       = job.find('li' , class_ = 'fleft br2 placeHolderLi experience').text.strip()
        except:
            experience       = np.nan

        try:
            salary           = job.find('li' , class_ = 'fleft br2 placeHolderLi salary').text.strip()
        except:
            salary           = np.nan

        try:
            location         = job.find('li' , class_ = 'fleft br2 placeHolderLi location').text.strip()
        except:
            location         = np.nan
        try:
            responsibilities = ','.join([i.text.strip() for i in (job.find('ul' , class_ = 'tags has-description').find_all('li'))])
        except:
            responsibilities = np.nan

        try:
            posted_on         = job.find('div' , class_ = 'tupleTagsContainer').text.strip()
        except:
            posted_on         = np.nan




        data.append([job_id, job_role, company, rating, reviews, responsibilities, posted_on , experience, location , salary , job_link , company_link])    
```

```python
df = pd.DataFrame(data, columns = ['job_id', 'job_role', 'company', 'rating', 'reviews','responsibilities', 'posted_on' , 'experience', 'location' , 'salary' , 'job_link' , 'company_link'])
```
